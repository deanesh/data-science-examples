````python
import numpy as np
from tensorflow import keras
from keras import layers

# Sample data
X = np.array([1000, 1500, 2000, 2500, 3000, 4500], dtype=float)
y = np.array([150000, 200000, 250000, 300000, 350000, 500000], dtype=float)

# Build model
model = keras.Sequential(
    [
        keras.Input(shape=(1,)),
        layers.Dense(64, activation="relu"),
        layers.Dense(32, activation="relu"),
        layers.Dense(1),
    ]
)

# Compile and train
model.compile(optimizer="adam", loss="mean_squared_error")
model.fit(X, y, epochs=500, verbose=0, validation_split=0.2)

# Predict for new input
input_sqft = 4000
predicted_price = model.predict(np.array([[input_sqft]]))

# Print predicted label (price)
print(f"üè† Predicted price for {input_sqft} sq ft: ${predicted_price[0][0]:,.2f}")
````

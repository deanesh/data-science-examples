````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")
print(tips.columns)
# Create a FacetGrid to facet by the 'day' column
g = sns.FacetGrid(tips, col="day")

# Map a histogram of 'total_bill' onto each facet
g.map(plt.hist, "total_bill", bins=7, color="lightblue", edgecolor="red")

# Show the plot
plt.show()
````

````python
sns.relplot(x="size", y="total_bill", data=tips)
plt.show()
````

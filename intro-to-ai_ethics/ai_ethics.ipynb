#### 2. **Fairness and Bias**

* Understand the difference between:

  * **Bias in data** (e.g., historical discrimination)
  * **Bias in models** (e.g., learned prejudice)
* Types of fairness:

  * **Group fairness**: Equal treatment for groups (e.g., gender, race)
  * **Individual fairness**: Similar people get similar outcomes

#### 4. **Accountability and Transparency**

* Who is responsible when AI fails or harms someone?
* The importance of:

  * **Explainability**: Making model decisions understandable
  * **Documentation**: Keeping records of data and modeling decisions
  * **Audits**: Regularly testing for bias and fairness
